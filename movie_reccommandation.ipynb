```python
movies=movies.merge(credits,on='title')
movies.head(2)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
  l=[]
  for i in ast.literal_eval(text):
    l.append(i['name'])
  return l
```

```python
movies.head()
```

```python
#movies['genres']=movies['genres'].apply(convert)
#movies['keywords'] = movies['keywords'].apply(convert)

#movies.head()
```

```python
def convert3(text):
  l=[]
  cnt=0
  for i in ast.literal_eval(text):
    if cnt<3:
      l.append(i['name'])
    cnt+=3
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(crew):
  l=[]
  for i in ast.literal_eval(crew):
    if i['job']=='Director':
      l.append(i['name'])
  return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def collapse(l):
  l1=[]
  for i in l:
     l1.append(i.replace(" ",""))
  return l1
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags']=movies['genres']+movies['overview']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies.drop(['genres','overview','keywords','cast','crew'],axis='columns')
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
vector=count.fit_transform(new_df['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(vector, vector)
```

```python
def get_recommendations(title):
    idx = new_df[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    movie_indices = [i[0] for i in sim_scores]
    return title.iloc[movie_indices]

```
